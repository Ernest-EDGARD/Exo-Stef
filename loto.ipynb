```python
"""
  - loto boule de 1 à 49
  - 5 numéros
  - 1 complementaire
  
  ex : 45 2 6 7 54 ... 7   
"""
import random

loto = range(1, 49)

numero = random.sample(loto, 5)
print(f"voici les numéros du loto {numero}")

complementaire = random.randint(1,9)
print(f"voici le numéro complémentaire {complementaire}")
  
```
